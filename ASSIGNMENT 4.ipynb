# Questions 2:
__How to create a series from a numpy array?__

```python
# creating numpy array
import numpy as np

arr = np.array([1,4,7,9,13])

print(" The Numpy array is :", arr)
```

```python
# creating series from numpy array

# importing pandas library
import pandas as pd

series = pd.Series(arr)

print("The Series is :")
print(series)

```

# Questions 3:
__How to convert the index of a series into a column of a dataframe?__


```python
# Creating Series 

# import the pandas library and aliasing as pd
import pandas as pd
import numpy as np
data = {'banana' : 10 , 'mango' : 24 , 'apple' : 14}
s = pd.Series(data)
print (s)

print("\nThe index of the series is :",s.index)
```

```python
# creating DataFrame
df = pd.DataFrame(s)

df
```

```python
# creating column from index of series
df['index']=df.index

df
```
